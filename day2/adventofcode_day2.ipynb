```python
import re
import unittest
```

```python
def parse_pkg(str):
    """Parses a package dimensions string into its dimensions tuple: 'lxwxh' -> (l, w, h)"""

    match = re.search('^(\d+)x(\d+)x(\d+)$', str)
    if match:
        return map(int, match.group(1, 2, 3))
    else:
        print 'Error parsing package dimensions: \'', str, '\'.'
```

```python
parse_pkg('1x2x3')
```

```python
class TestParsePkg(unittest.TestCase):
    def test_empty(self):
        self.assertEqual(parse_pkg(''), None)

    def test_incomplete1(self):
        self.assertEqual(parse_pkg('1x2'), None)

    def test_incomplete2(self):
        self.assertEqual(parse_pkg('1x2x'), None)

    def test_whitespace1(self):
        self.assertEqual(parse_pkg(' 1x2x3'), None)

    def test_whitespace2(self):
        self.assertEqual(parse_pkg(' 1x 2 x 3'), None)

    def test_pos1(self):
        self.assertEqual(parse_pkg('1x2x3'), [1, 2, 3])

    def test_pos2(self):
        self.assertEqual(parse_pkg('11x204x30045'), [11, 204, 30045])

    def test_neg1(self):
        self.assertEqual(parse_pkg('-11x2x3'), None)

suite = unittest.TestLoader().loadTestsFromTestCase(TestParsePkg)
unittest.TextTestRunner(verbosity=3).run(suite)
```
